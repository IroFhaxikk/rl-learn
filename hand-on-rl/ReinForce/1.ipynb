```python
from ReinForce import PolicyNet
import torch
import time
def visualize_trained_policy(model_path, episodes=5):
    env = gym.make("CartPole-v1", render_mode="human")

    input_dim = env.observation_space.shape[0]
    action_dim = env.action_space.n

    # 创建网络并加载权重
    policy = PolicyNet(input_dim, 256, action_dim)
    policy.load_state_dict(torch.load(model_path, map_location="cpu"))
    policy.eval()

    print("Model loaded OK. Start playing...")

    for ep in range(episodes):
        state, _ = env.reset()
        done = False
        total_reward = 0

        while not done:
            # 环境渲染
            env.render()

            # 模型决定动作
            s = torch.tensor([state], dtype=torch.float)
            with torch.no_grad():
                probs = policy(s)
                action = torch.argmax(probs, dim=1).item()

            # 执行动作
            next_state, reward, done, truncated, info = env.step(action)
            state = next_state
            total_reward += reward

            time.sleep(0.02)  # 放慢模拟速度，看得清楚一点

        print(f"Episode {ep+1}: Total reward = {total_reward}")

    env.close()
```
